# Data Merging

```python
import pandas as pd
import numpy as np
```

```python
# create a dataset with 20 rows and 3 columns   

df1 = pd.DataFrame({'Name': ['John', 'Jane', 'Jack', 'Jill', 'Joe', 'Jenny','Jim', 'Jenny', 'Joe', 'Jenny', 'Joe'],
                   'Score': [80, 90, 95, 100, 90, 85, 80, 75, 70, 65, 60]})
df2 = pd.DataFrame({'Age': [20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
                    'Score': [80, 90, 95, 100, 90, 85, 80, 75, 70, 65, 60]})
```

```python
df1
```

```python
df2
```

```python
# merge datasets
merged_df = pd.merge(df1, df2, on='Score')
merged_df
```

# Data Concatinating

```python
# create a dataset with 20 rows and 3 columns   

data1 = pd.DataFrame({'Name': ['John', 'Jane', 'Jack', 'Jill', 'Joe', 'Jenny','Jim', 'Jenny', 'Joe', 'Jenny', 'Joe'],
                   'Score': [80, 90, 95, 100, 90, 85, 80, 75, 70, 65, 60]})
data2 = pd.DataFrame({'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose', 'Austin'],
                    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]})
```

```python
# concatinate the datasets
concatenated = pd.concat([data1, data2], axis=1)
concatenated
```
