# Feature Encoding

Feature encoding is the process of converting categorical variables into numerical variables. This is done to make the data more suitable for machine learning algorithms. The categorical variables are the ones that have a limited number of categories. For example, the gender of a person can be either male or female, and the day of the week can be either Sunday, Monday, Tuesday, Wednesday, Thursday, Friday, or Saturday. These variables are called categorical variables because they have a limited number of categories.

Tyoes or label encoding are as under:-
1. Ordinal Encoding
2. One Hot Encoding
3. Label Encoding

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
df['encoded_time'].value_counts()
```

#### Here 0 is Dinner and 1 is Lunch

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
df['encoded_day'].value_counts()
```

#### similarily here 1,2,3,4 are Thursday, Friday, Saturday and Sunday respectively.
